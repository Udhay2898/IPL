```python
for column in df.columns:
  unique_value = df[column].unique()
  print(f"{column} :\n {unique_value}\n")
```

```python
df.head(1)
```

```python
df = df[df['Current Team(s)'] != 'Retired']
```

```python
team_gr = df.groupby(['Current Team(s)','Batting Hand']).agg({'Matches' : 'sum',
                                            'Innings' : 'sum',
                                            'Average' : 'mean',
                                             'Runs' : 'sum',
                                             'Balls Faced' : 'sum',
                                             'Strike Rate' : 'mean',
                                             '100' : 'sum',
                                             '50'	: 'sum',
                                             'Duck Outs' : 'sum',
                                             'NO' : 'sum',
                                             '4s' : 'sum',
                                             '6s' : 'sum'}).reset_index()

team_gr.head(2)
```

**Team Performance Analysis**

**Matches and Innings by Each Teams**

```python
gr_1= df.groupby('Player').agg({'Matches' : 'sum',
                                'Innings' : 'sum'})

gr_1.plot.line(figsize = (20,15), marker = '*',grid = True)

tick = range(len(gr_1))
plt.xticks(ticks = tick, labels = gr_1.index, rotation = 90)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.title("Matches and Innings by Batting Hand for Each Teams")

plt.show()
```

**Team Performance Analysis by Batting Hand**

**Matches and Innings by Batting Hand for Each Teams**

**Line Chart**

```python
pivot_team_gr = team_gr.pivot(index = 'Current Team(s)',columns = 'Batting Hand', values = ['Matches',	'Innings'] )

pivot_team_gr.plot.line(figsize = (10,5), marker = '*',grid = True)

tick = range(len(pivot_team_gr))
plt.xticks(ticks = tick, labels = pivot_team_gr.index, rotation = 0)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.title("Matches and Innings by Batting Hand for Each Teams")

plt.show()
```

**BAR Chart**

```python
pivot_team_gr.plot.bar(figsize = (10,5),grid = True)

tick = range(len(pivot_team_gr))
plt.xticks(ticks = tick, labels = pivot_team_gr.index, rotation = 0)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.title("Matches and Innings by Batting Hand for Each Teams")
plt.show()
```

**PIE Chart**

```python
explode = [0.01] * len(pivot_team_gr)

fig, axes = plt.subplots(nrows=1, ncols=len(pivot_team_gr.columns), figsize=(20, 10))
pivot_team_gr.plot.pie(subplots=True, ax=axes, legend=False, autopct='%1.1f%%', startangle = 90,explode = explode)

plt.title("Matches and Innings by Batting Hand for Each Teams")
plt.tight_layout()

plt.show()
```
