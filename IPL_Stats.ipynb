```python
player_count = df[df['Current Team(s)'] != 'Retired' ]
player_count = player_count.groupby('Current Team(s)').agg({'Player' : 'count'}).reset_index()

player_count.columns = ['Team(s)','Player_Count']
player_count
```

```python
ticks = range(len(player_count))
player_count.plot(figsize = (10,8),marker = 'o',grid = True,legend = False)

plt.xticks(ticks=ticks, labels=player_count['Team(s)'], rotation=0)

plt.show()
```

```python
_6s_hit = df[df['Current Team(s)'] != 'Retired']

_6s_hit = _6s_hit.groupby(['Current Team(s)','Batting Hand']).agg({'6s' : 'sum',
                                                                   '4s' : 'sum'}).reset_index()

_6s_hit.head(2)
```

```python
_6s_hit_pivot = _6s_hit.pivot(index = 'Current Team(s)', columns = 'Batting Hand', values = ['6s','4s'])

_6s_hit_pivot.head(2)
```

```python
_6s_hit_pivot.plot(figsize = (10,6), marker = 'o',grid = True)

ticks = range(len(_6s_hit_pivot))
plt.xticks(ticks=ticks, labels = _6s_hit_pivot.index, rotation=0)

plt.xlabel('Team Name')
plt.ylabel('Number of 6s')
plt.title('Number of 6s by Team')
plt.grid(True)
plt.show()
```
