```python
group_data = df[df['Current Team(s)'] != 'Retired']
group_data = group_data.groupby(['Current Team(s)','Batting Hand']).agg({'100' : 'sum',
                                                                         '50' : 'sum',
                                                                         'Duck Outs' : 'sum'}).reset_index()
group_data
```

```python
pivot_group_data = group_data.pivot(index = 'Current Team(s)',columns = 'Batting Hand',values=['100','50','Duck Outs'])
pivot_group_data
```

```python
pivot_table_1 = df.pivot_table(index='Current Team(s)', columns='Batting Hand', values=['Matches', 'Runs', '100', 'Duck Outs'], aggfunc='sum')
pivot_table_1
```

```python
pivot_table_2 = df.pivot_table(index='Player', columns='Played in 2024', values=['Matches', 'Runs', 'High Score'], aggfunc='sum')
pivot_table_2
```
